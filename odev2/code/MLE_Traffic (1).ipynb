```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

## Veri Yükleme

```python
traffic_data = np.array([12, 15, 10, 8, 14, 11, 13, 16, 9, 12, 11, 14, 10, 15])
print('Traffic data:', traffic_data)
```

## Negative Log Likelihood

```python
def negative_log_likelihood(lam, data):
    lam = lam[0] if isinstance(lam, (list, np.ndarray)) else lam
    log_likelihood = np.sum(data * np.log(lam) - lam)
    return -log_likelihood
```

## Sayısal MLE

```python
initial_guess = [1.0]
result = opt.minimize(negative_log_likelihood, initial_guess, args=(traffic_data,), bounds=[(0.001, None)])
lambda_mle = result.x[0]
print('Numerical MLE lambda:', lambda_mle)
print('Analytical lambda (mean):', np.mean(traffic_data))
```

## Model Görselleştirme

```python
plt.hist(traffic_data, bins=8, density=True, alpha=0.6, label='Observed Data')
x = np.arange(0, 25)
pmf = poisson.pmf(x, lambda_mle)
plt.plot(x, pmf, 'o-', label=f'Poisson PMF (λ={lambda_mle:.2f})')
plt.xlabel('Vehicles per minute')
plt.ylabel('Probability')
plt.title('Poisson Model Fit to Traffic Data')
plt.legend()
plt.show()
```

## Outlier Analizi

```python
traffic_outlier = np.append(traffic_data, 200)
print('New mean with outlier:', np.mean(traffic_outlier))
```
